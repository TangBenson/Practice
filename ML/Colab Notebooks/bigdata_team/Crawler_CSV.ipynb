```python
def download(s,y,m,d,utf8,high):
    url=f"https://e-service.cwb.gov.tw/HistoryDataQuery/DayDataController.do?command=viewMain&station="\
        f"{s}&stname={utf8}&datepicker={y}-{str(m).zfill(2)}-{str(d).zfill(2)}&altitude={high}m"  #網址
    csvfile=f"{s}-{y}-{str(m).zfill(2)}-{str(d).zfill(2)}.csv" #開個csv檔案準備寫入
    r=requests.get(url)
    r.encoding="utf-8"
    soup=BeautifulSoup(r.text,"lxml")
    tag_table=soup.find(id="MyTable") #用BeautifulSoup找到table位置
    rows=tag_table.findAll("tr") #找到每個
    check = soup.findAll("label", text="本段時間區間內無觀測資料！")       
    if check==[]:
        #'r:讀 w:寫 w+:讀寫(存在則覆蓋,不存在則新建)'
        #開頭+with就不擔心忘記close()了
        with open("./Temp3Hr/"+csvfile,"w+",newline="",encoding="utf-8") as fp:
            writer=csv.writer(fp)
            for row in rows:
                rowList=[]
                for cell in row.findAll(["td","th"]):
                    rowList.append(cell.get_text().replace("\n","").replace("\r",""))
                writer.writerow(rowList)
```

```python
# 抓每日
import requests
from bs4 import BeautifulSoup
import csv
import time

StationList = list(df_station['站號'])
YearList=['2022']
MonthList = [1]

for y in YearList:
    time.sleep(1)
    for s in StationList:
        for m in MonthList:
            for d in range(1,32):
                try:
                    utf8 = list(df_station[df_station['站號']==s]['utf8'])[0]
                    high = list(df_station[df_station['站號']==s]['海拔高度(m)'])[0]
                    download(s,y,m,d,utf8,high)
                except:
                    continue
```
