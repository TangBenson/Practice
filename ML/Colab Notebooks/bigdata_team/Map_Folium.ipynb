### 熱圖範例

### 熱度圖，顯示連續的時間軸 (完全不懂意思)

#### 输入HeatMapWithTime中的data_move是一个多层的list：[[[Y1,X1,W1],...],[[Y1,X1,W1],...],...,[[Y1,X1,W1],...]] 需要按照规范进行构造

```python
from folium.plugins import HeatMapWithTime

center_pos = [35.712326, 139.804037]

# 使用 numpy 建立初始資料
initial_data = (np.random.normal(size=(200, 2)) * # =>array([[-0.89878316, -0.34068839],[-0.36938624,  1.82686894]])
                np.array([[0.02, 0.02]]) +        # =>array([[ 0.00323325, -0.01993604],[ 0.02237776, -0.01660395]])
                np.array([center_pos]))           # =>array([[ 35.6942639 , 139.82209956],[ 35.66657485, 139.78028804]])

# 建立連續資料
data = [initial_data.tolist()] # =>[[[35.6853127304945, 139.77823553948116],[35.686892241268744, 139.81479795318185]]]
# print(len(data))        1
# print(len(data[0]))     200
# print(len(data[0][0]))  2

for i in range(20): #20表示熱力圖的橫軸有20個可以選，第1~21筆資料的熱度圖
    data.append((data[i] + np.random.normal(size=(200, 2)) * 0.001).tolist())
# print(len(data))        21
# print(len(data[0]))     200
# print(len(data[0][0]))  2

print(data[0][0])
print(data[0][1])
print(data[0][0][0])
print(data[0][0][1])
print(data[0][1][0])
print(data[0][1][1])

fmap = folium.Map(center_pos, zoom_start=11)
fmap.add_child(HeatMapWithTime(data)) # 顯示連續熱度圖
```
