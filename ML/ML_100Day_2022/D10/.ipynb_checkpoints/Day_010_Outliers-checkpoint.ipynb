```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = 'data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

df.head()
```

```python
df_train['SalePrice']
```

```python
train_Y = np.log1p(df_train['SalePrice'])
train_Y
```

```python
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
list(zip(df.dtypes, df.columns))[0]
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
## 過濾出所有數值型欄位
df = df[num_features]
df = df.fillna(-1) # 為何NaN要填入-1？
MMEncoder = MinMaxScaler()
df.head()
```

```python
# 顯示 GrLivArea 與目標值的散佈圖
## GrLivArea: Above grade (ground) living area square feet
## sns.regplot: https://seaborn.pydata.org/generated/seaborn.regplot.html
## https://www.cnblogs.com/cgmcoding/p/13293395.html
import seaborn as sns
import matplotlib.pyplot as plt
print(df['GrLivArea'])
sns.regplot(x = df['GrLivArea'], y=train_Y)
plt.show();
```

```python
# 做線性迴歸, 觀察分數
## 做Min-Max Scaler 處理
# print(df)
train_X = MMEncoder.fit_transform(df)
# print(train_X)
estimator = LinearRegression()
## https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
# cross_val_score會把資料丟進estimator跑？不用我先自己跑estimator.fit?
```

```python
# 將 GrLivArea 限制在 800 到 2500 以內, 調整離群值
## .clip: 大於或小於閾值的值，以閾值取代
df['GrLivArea'] = df['GrLivArea'].clip(800, 2500)
sns.regplot(x = df['GrLivArea'], y=train_Y)
plt.show()
```

```python
# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 GrLivArea 限制在 800 到 2500 以內, 捨棄離群值
keep_indexs = (df['GrLivArea']> 800) & (df['GrLivArea']< 2500)
df = df[keep_indexs]
train_Y = train_Y[keep_indexs]
sns.regplot(x = df['GrLivArea'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```
